```python
import pandas as pd, numpy as np
```

```python
# Intiializes data
data = pd.read_csv("clint.csv").drop("Unnamed: 0", 1)
stop = open("stop-word-list.txt").read().splitlines()
d = data[data.message.notnull()]
```

```python
# Removes the punctuations from the data set
punctuation_chars = ".,!?:;\"'-~|[]()/\\\n"

messages = d.message.str.lower()
def remove_punctuations(s):
    s = ''.join([i for i in s if i not in frozenset(punctuation_chars)])
    return s
messages = messages.apply(remove_punctuations)
messages
```

```python
# Removes multiple instances of the same message boxes
uniques = set(messages)
uniques
```

```python
# Prints the length of the unique messages
print(len(uniques))
```
